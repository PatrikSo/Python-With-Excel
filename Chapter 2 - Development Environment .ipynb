# Chapter 2 - Development Environment 

- Can write VBA or Power Queries on Excel alone.
- Using Anaconda Python and Jupyter notebook.
- Brief Descriptions of the development enviroments used:

| Anaconda (sum.) | A scientific Python distribution |
| --- | --- |
| Python | Language at the core of Anaconda. Python interpreter. |
| Anaconda Prompt | Command Prompt/Terminal, uses proper Python interpreter and packages | 
| Jupyter notebook, pandas, xlwings, so on | Many third-party packages pre-installed |
| Conda & pip | Package managers allowing package installations and updating existing ones |


| VS Code (sum) | Powerful text editor |
| --- | --- |
| Python extension | Allows you to run Pthon code and Jupyter notebooks |
| Debugger | setting breakpoints and allows setting up steps through lines of code |
| Terminal | Integrated terminal used as an Anaconda prompt |
| Git | Integrated with Git VCS |

- Pip commands for terminal
| Action | Pip |
| --- | --- |
| List all installed packages | pip freeze |
| Install the latest package version | pip install package |
| Install a specific package version | pip install package==1.0.0 |
| Update a package | pip install --upgrade package |
| Uninstall a package | pip uninstall package |

- Jupyter pages fun!
- Jupyter page shortcuts/commands
| Shortcut/Command | Action |
| --- | --- |
| Shift+Enter | run the cell (works also in edit mode) |
| up arrow | move cell selector up |
| down arrow | move cell selector down |
| b | insert a new cell below the current cell |
| a | insert a new cell above the current cell |
| dd | Delete the current cell (two d's, not a typo!) |
| m | change cell type to markdown |
| y | change cell type to code
